```python
import numpy as np
import cv2 
from shapely.geometry import Polygon, Point
import matplotlib.pyplot as plt

# Making The Blank Image
image = np.zeros((500,500,3))
drawing = False
ix = 0
iy = 0
px, py = None, None
path = []
all_poly = []
# Adding Function Attached To Mouse Callback
def draw(event,x,y,flags,params):
    global ix,iy,drawing,path,px,py
    # Left Mouse Button Down Pressed
    # print(x%50,y%50 )
    # check if mouse left button up
    if event==4:
        ix = x
        iy = y
        # print("clicked ", (ix, iy))
        if px is not None and py is not None:
            if abs(px-ix) > abs(py-iy):
                # print("# hozizontal move")
                iy = py
            elif abs(px-ix) < abs(py-iy):
                # print("# vertical move")
                ix = px
            cv2.line(image,pt1=(px,py),pt2=(ix,iy),color=(255,255,255),thickness=1)
        path.append((ix,iy))
        px, py = ix, iy
        

# Making Window For The Image
cv2.namedWindow("Window")

# Adding Mouse CallBack Event
cv2.setMouseCallback("Window",draw)

# Starting The Loop So Image Can Be Shown
while(True):

    cv2.imshow("Window",image)

    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
    # if n is pressed
    if cv2.waitKey(20) & 0xFF == ord('n'):
        image = np.zeros((500,500,3))
        path[0] = path[-1]
        print(path)
        poly = Polygon(path)
        x, y = poly.exterior.xy
        plt.plot(x, y, c="red")
        plt.show()        
        all_poly.append(path)
        path = []
        print("New Image")
cv2.destroyAllWindows()

print(path)
```
