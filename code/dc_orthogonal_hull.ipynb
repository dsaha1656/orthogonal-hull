```python
# Create figure and axes
fig, ax = plt.subplots(figsize = (10,100))

# Display the image
ax.imshow(img)
ax.set_figure = (10,100)

# Create a Rectangle patch
rect = patches.Rectangle((500, 100), 10, 10, linewidth=5, edgecolor='r', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)
```

```python
def markPoint(x, y):
    # Create a Rectangle patch
    rect = patches.Rectangle((x, y), 3, 3, linewidth=5, edgecolor='r', facecolor='none')
    return rect
```

```python
# Create figure and axes
fig, ax = plt.subplots(figsize = (10,100))
x_loop_must_break = False
points = []
for col_num, col in enumerate(img):
    for row_num, row in enumerate(col):
        if(row==0):
            points.append( (row_num, col_num) )
            # x_loop_must_break = True
            # break 
    if x_loop_must_break: break

for point in points:
    # print(point[0],point[1])
    # break
    ax.add_patch(markPoint(point[0],point[1]))

# Add the patch to the Axes
# ax.add_patch(markPoint(100,200))
# Display the image
ax.imshow(img)
ax.set_figure = (10,100)
```
